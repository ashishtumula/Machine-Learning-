```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/jaysetty/machine_learning/master/Cust_Spend_Data.csv")
```

```python
X = data.drop(['Cust_ID', 'Name'], axis=1)
```

### WITHOUT SCALING

#### a) Euclidean Distance

```python
# Distance Computation

from scipy.spatial.distance import pdist, squareform

d_euc = pdist(X, metric="euclidean")


# Building Clusters 

from scipy.cluster.hierarchy import linkage, dendrogram, cut_tree

clus_euc = linkage(d_euc, method = "average")
```

```python
# Displaying the clusters in Dendrogram

import matplotlib.pyplot as plt
dendrogram(clus_euc, labels=data.Name.values.tolist())
plt.xlabel('hclust')
plt.ylabel('Euclidean Distance')
plt.suptitle('Cluster Dendrogram(Euclidean)', fontweight='bold', fontsize='14')
```

#### b) Chebyshev Distance

```python
d_che = pdist(X, metric="chebyshev")

clus_che = linkage(d_che, method = "average")
```

```python
dendrogram(clus_che, labels=data.Name.values.tolist())
plt.xlabel('hclust')
plt.ylabel('Chebyshev Distance')
plt.suptitle('Cluster Dendrogram(Chebyshev)', fontweight='bold', fontsize='14')
```

### WITH SCALING

```python
from scipy.stats import zscore
X_z = X.apply(zscore)
```

#### a) Eucldean Distance

```python
d_euc_scaling = pdist(X_z, metric="euclidean")

clus_euc_scaling = linkage(d_euc_scaling, method = "average")
```

```python
dendrogram(clus_euc_scaling, labels=data.Name.values.tolist())
plt.xlabel('hclust_scaling')
plt.ylabel('Euclidean Distance')
plt.suptitle('Cluster Dendrogram(Euclidean) with Scaling', fontweight='bold', fontsize='14')
```

#### b) Chebyshev Distance

```python
d_che_scaling = pdist(X_z, metric="chebyshev")

clus_che_scaling = linkage(d_che_scaling, method = "average")
```

```python
dendrogram(clus_che_scaling, labels=data.Name.values.tolist())
plt.xlabel('hclust_scaling')
plt.ylabel('Chebyshev Distance')
plt.suptitle('Cluster Dendrogram(Chebyshev) with Scaling', fontweight='bold', fontsize='14')
```
